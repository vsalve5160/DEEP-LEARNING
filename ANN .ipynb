# Data Exploration and Preprocessing

```python
import numpy as np

seed = 7
np.random.seed(seed)
X = df.drop(columns = ['letter'])
y = df['letter']
```

```python
Xtrain,Xtest,ytrain,ytest = train_test_split(X,y,test_size = 0.20, random_state=1)
```

```python
le = LabelEncoder()
ytrain_encoded_labels = le.fit_transform(ytrain)
ytest_encoded_labels = le.transform(ytest)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
a = StandardScaler()
a.fit(X)
X_standardized = a.transform(X)
```

# MODEL IMPLIMENTATIN

```python
model = Sequential()
model.add(Dense(64, input_dim=16,  activation='relu')) #1st layer
model.add(Dense(34,  activation='relu')) #2nd layer
model.add(Dense(26, activation='softmax')) #3rd layer or op layer
```

```python
df['letter'].nunique()
```

```python
model.compile(loss='Categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
ytrain_encoded = to_categorical(ytrain_encoded_labels)
ytest_encoded = to_categorical(ytest_encoded_labels)
```

```python
model.compile(
    loss='categorical_crossentropy',   # for multi-class classification
    optimizer='adam',
    metrics=['accuracy']
)
```

```python
history = model.fit(Xtrain, ytrain_encoded, epochs=10, batch_size=32, validation_split=0.2)
```

```python
print(Xtrain.shape)        # should be (rows, 16)
print(ytrain_encoded.shape)  # should be (rows, 26)
```

```python
# Evaluate on test set
test_loss, test_acc = model.evaluate(Xtest, ytest_encoded)
print(f"Test Accuracy: {test_acc:.4f}")
```

```python
from sklearn.metrics import classification_report
import numpy as np

# Predict classes
y_pred_probs = model.predict(Xtest)             # probabilities
y_pred_classes = np.argmax(y_pred_probs, axis=1) # predicted labels
y_true = np.argmax(ytest_encoded, axis=1)        # true labels

# Classification report
print(classification_report(y_true, y_pred_classes))
```

```python
# --- Step 1: Imports ---
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score, classification_report
import numpy as np

# --- Step 2: Build model function ---
def create_model(learning_rate=0.001, activation='relu', neurons=32):
    model = Sequential()
    model.add(Dense(neurons, input_dim=Xtrain.shape[1], activation=activation))
    model.add(Dense(neurons, activation=activation))
    model.add(Dense(ytrain_encoded.shape[1], activation='softmax'))  # for multi-class classification
    optimizer = Adam(learning_rate=learning_rate)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['accuracy'])
    return model

# --- Step 3: Wrap model ---
model = KerasClassifier(model=create_model, verbose=0)

# --- Step 4: Define parameter grid ---
param_grid = {
    "batch_size": [16, 32],
    "epochs": [10, 20],
    "model__learning_rate": [0.001, 0.01],
    "model__activation": ["relu", "tanh"],
    "model__neurons": [16, 32]
}

# --- Step 5: Randomized Search (faster than GridSearchCV) ---
random_search = RandomizedSearchCV(
    estimator=model,
    param_distributions=param_grid,
    n_iter=5,   # test only 5 random combinations
    cv=3,
    n_jobs=-1,
    verbose=1
)

# --- Step 6: Fit the model ---
random_result = random_search.fit(Xtrain, ytrain_encoded)

# --- Step 7: Display best results ---
print("✅ Best Parameters Found:", random_result.best_params_)
print("✅ Best Cross-Validation Accuracy:", random_result.best_score_)
```

# Evaluate Tuned Model

```python
# Evaluate tuned model
best_model = random_result.best_estimator_
y_pred_tuned = best_model.predict(Xtest)

# Convert predictions back to class labels if one-hot encoded
y_pred_tuned = np.argmax(y_pred_tuned, axis=1)
y_true = np.argmax(ytest_encoded, axis=1)

print("✅ Accuracy (Tuned Model):", accuracy_score(y_true, y_pred_tuned))
print("\n✅ Classification Report:\n", classification_report(y_true, y_pred_tuned))
```
